# Import text file and split into sentences

```python
import re
import spacy
from spacy.lang.en import English
from spacy.cli import download

from pathlib import Path

path = "./test.txt"
text  = open(path, "r").read()

nlp = English()
nlp.add_pipe("sentencizer")
doc = nlp(text)

def unmatch_parentheses(text):
    starts = len(re.findall('\(', text))
    ends   = len(re.findall('\)', text))

    return starts != ends

sentences = []
curr = prev = None
for sentence in list(doc.sents):
    curr = str(sentence)
    
    if prev == None:
        next
    else:
        if curr.endswith(').') and unmatch_parentheses(curr) and unmatch_parentheses(prev):
            del sentences[-1]
            sentences.append(prev + curr)
        else:
            sentences.append(curr)
    prev = curr
    
print("%d sentences detected" % len(sentences))
```

# Filter by keywords

```python
keywords = ['induce', 'silence', 'promote',  'inhibit',  'decrease',  'increase', 'upregulate']

nlp = spacy.load("en_core_web_trf")

docs = []
docs = list(map(lambda sentence: nlp(sentence), sentences))
key_sentences = []

for doc in docs:
    for token in doc:
        if token.pos_ == "VERB" and token.lemma_ in keywords:
            key_sentences.append(doc)
```

```python
from spacy import displacy
sample = key_sentences[0]
displacy.render(sample, style="dep")
```

```python
for token in sample:
    print(token.text, token.pos_, token.dep_, token.head.text, token.head.pos_, [child for child in token.children])
```
